**Detect Outliers Using IQR**

```python
import pandas as pd

# Detect outliers using IQR
def detect_outliers_iqr(df):
    # Select only the numeric columns
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns

    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[num_cols].quantile(0.25)
    Q3 = df[num_cols].quantile(0.75)
    IQR = Q3 - Q1

    # Set the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return lower_bound, upper_bound

# Calculate the bounds for outliers
lower_bound, upper_bound = detect_outliers_iqr(df)
print(f"Lower bounds: \n{lower_bound}")
print(f"Upper bounds: \n{upper_bound}")
```

**Handling Outliers**

```python
import numpy as np

# Cap outliers using the calculated lower and upper bounds
def cap_outliers(df, lower_bound, upper_bound):
    # Select only numeric columns
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns

    # Apply capping to replace outliers with the boundary values
    df[num_cols] = df[num_cols].apply(lambda col: np.clip(col, lower_bound[col.name], upper_bound[col.name]))

    return df

# Cap the outliers
df = cap_outliers(df, lower_bound, upper_bound)

# Check the shape of the dataset after handling outliers
print(f"Shape of dataset after handling outliers: {df.shape}")
```

**Verifying Outliers Are Handled**

```python
# Verify if there are any remaining outliers after capping
def check_remaining_outliers(df, lower_bound, upper_bound):
    # Select numeric columns
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns

    # Check for values below the lower bound and above the upper bound
    outliers_below = (df[num_cols] < lower_bound).sum()
    outliers_above = (df[num_cols] > upper_bound).sum()

    print(f"Outliers below lower bound: \n{outliers_below[outliers_below > 0]}")
    print(f"Outliers above upper bound: \n{outliers_above[outliers_above > 0]}")

# Verify the result after capping
check_remaining_outliers(df, lower_bound, upper_bound)
```

# **PCA (Principal Component Analysis) Implementation**

**Standardize the Data**

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the numerical features
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[num_cols])
```

**Applying PCA**

```python
# Initialize PCA and fit to the scaled data
pca = PCA(n_components=0.95)  # Keep 95% of the variance
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(data=pca_result, columns=[f'PC{i + 1}' for i in range(pca_result.shape[1])])
```

**Variance Explained by Each Principal Component**

```python
# Plot the explained variance ratio
plt.figure(figsize=(14, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(ticks=range(1, len(pca.explained_variance_ratio_) + 1))
plt.grid()
plt.show()
```

**Visualize the First Two Principal Components**

```python
# Scatter plot of the first two principal components
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_df['PC1'], y=pca_df['PC2'], alpha=0.5)
plt.title('PCA Result: First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()
```

**Analyze PCA Loadings**

```python
# Create a DataFrame for PCA loadings
loadings = pd.DataFrame(pca.components_.T, index=num_cols, columns=[f'PC{i + 1}' for i in range(pca_result.shape[1])])

# Display the top features contributing to the first two principal components
top_features_pc1 = loadings['PC1'].nlargest(15)
top_features_pc2 = loadings['PC2'].nlargest(15)

print("Top 15 Features for Principal Component 1:")
print(top_features_pc1)
print("\nTop 15 Features for Principal Component 2:")
print(top_features_pc2)
```

**Visualize PCA Loadings**

```python
# Plot the loadings of the first two principal components
plt.figure(figsize=(12, 6))
plt.barh(top_features_pc1.index, top_features_pc1.values, color='skyblue')
plt.title('Top Features for Principal Component 1')
plt.xlabel('Loading Value')
plt.show()

plt.figure(figsize=(12, 6))
plt.barh(top_features_pc2.index, top_features_pc2.values, color='salmon')
plt.title('Top Features for Principal Component 2')
plt.xlabel('Loading Value')
plt.show()
```
